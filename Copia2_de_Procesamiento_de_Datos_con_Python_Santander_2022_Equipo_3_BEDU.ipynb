# 4. An√°lisis Exploratorio de datos üîé

A continuaci√≥n se realizar√° la lectura de los datos y se almacenar√°n en forma de dataframes en los objeto train y test, respectivamente:

```python
import pandas as pd
train = pd.read_excel(datos_train.content)
train
```

```python
test = pd.read_excel(datos_test.content)
test
```

Como puede observarse, los nombres de las columnas de los dataframes son incosistentes, por lo que se modificar√°n los nombres del conjunto train para que coincidan con los del conjunto test:

```python
column_name_mapping = {
    "ID": "Id",
    "CATEGORY": "Category",
    "SOURCE": "Source",
    "TOPICS": "Topic",
    "HEADLINE": "Headline",
    "TEXT": "Text",
    "LINK": "Link"
}
```

```python
test_renamed = test.rename(columns=column_name_mapping)
test_renamed 
```

Con el fin de desarrollar un an√°lisis m√°s completo, se concatenar√°n ambos dataframes, con el prop√≥sito de conseguir un marco que proporcione mayor cantidad de datos:

```python
noticias = pd.concat([train, test_renamed], axis=0)
noticias
```

```python
shape = noticias.shape
print(f"Las dimensiones del dataframe son: {shape}") # Interpolaci√≥n de strings
```

```python
# Importaci√≥n de librer√≠as 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk # importar toolkit del lenguaje natural
nltk.download('omw-1.4')
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords # importar palabras vac√≠as
import re # importar re para expresiones regulares
from collections import Counter # importar contador
from pandas._libs.tslibs import dtypes
```

```python
# Construir lematizador para identificar ra√≠ces de palabras en espa√±ol
wnl = nltk.stem.WordNetLemmatizer()

# Construir diccionario de palabras vac√≠as en espa√±ol
stop_words_es = stopwords.words('spanish')
stop_words_es.extend(("number","amlo","m√©xico","gobierno")) # Se incluyen en el diccionario como palabras vac√≠as porque por su alta frecuencia no aportan valor
stopwords_dict_es = Counter(stop_words_es)
```

```python
# Definir funci√≥n para procesamiento:
def nltk_proceso(text):
    wordlist = re.sub(r'[^\w\s]', '', text).split()
    text = ' '.join([wnl.lemmatize(word) for word in wordlist if word not in stopwords_dict_es])
    return  text
```

```python
# Procesar textos de las noticias
noticias_dropped["encabezado"] = noticias_dropped.encabezado.apply(nltk_proceso)

# Preprocesar encabezados de las noticias
noticias_dropped["texto"] = noticias_dropped.texto.apply(nltk_proceso)
```

```python
# Funci√≥n para n-gramas
def plot_top_ngrams(corpus, title, ylabel, xlabel="N√∫mero de ocurrencias", n=2):
  true_b = (pd.Series(nltk.ngrams(corpus.split(), n)).value_counts())[:20]
  true_b.sort_values().plot.barh(color="crimson", width=.9, figsize=(12, 8))
  plt.title(title)
  plt.ylabel(ylabel)
  plt.xlabel(xlabel)
  plt.show()
```

```python
stopwords = ["para", "en", "un", "el", "de", "y", "dentro", "del", "a", "desde","rt"]

def clean_tweet(tweet):
    if type(tweet) == np.float:
        return ""
    temp = tweet.lower()
    temp = re.sub("'", "", temp) # Para no remover contracciones en ingl√©s
    temp = re.sub("@[A-Za-z0-9_]+","", temp)
    temp = re.sub("#[A-Za-z0-9_]+","", temp)
    temp = re.sub(r'http\S+', '', temp)
    temp = re.sub('[()!?]', ' ', temp)
    temp = re.sub('\[.*?\]',' ', temp)
    temp = re.sub("[^a-z0-9]"," ", temp)
    temp = temp.split()
    temp = [w for w in temp if not w in stopwords]
    temp = " ".join(word for word in temp)
    temp=re.sub(r"([^n\u0300-\u036f]|n(?!\u0303(?![\u0300-\u036f])))[\u0300-\u036f]+", r"\1", normalize( "NFD", temp), 0, re.I)
    return temp
```
